```python
import sqlite3
####################################################################  
class Artist: 
    all_artists=[]
    
    def __init__(self,artist_id,name):
        self.id=artist_id
        self.name=name
        
    @classmethod
    def loads_all_artists(cls,cursor):
        cursor.execute('''SELECT ArtistId, Name FROM artists''')
        result=cursor.fetchall()
        
        for record in result:
            artist=Artist(record[0],record[1])
            cls.all_artists.append(artist)
            
        
    @classmethod
    def get_artist_object_from_id(cls,artist_id):
        for artist in cls.all_artists:
            if artist.id==artist_id:
                return artist
            
    @classmethod
    def display_all_artist(cls):
        for artist in cls.all_artists:
            print(artist.name)
####################################################################  
class Album:
    all_albums=[]
    
    def __init__(self,album_id,title,artist):
        self.id=album_id
        self.title=title
        self.artist=artist
        
    @classmethod
    def loads_all_albums(cls,cursor):
        cursor.execute('''SELECT AlbumId, Title, ArtistId FROM albums''')
        result=cursor.fetchall()
        
        for record in result:
            artist=Artist.get_artist_object_from_id(record[2])
            album=Album(record[0],record[1],artist)
            cls.all_albums.append(album)
            
    
    @classmethod
    def get_album_object_from_id(cls,album_id):
        for album in cls.all_albums:
            if album.id==album_id:
                return album
        print("get_album_object_from_id: Object not found for id {}".format(album_id))
        return None
            
    @classmethod
    def display_all_albums(cls):
        for album in cls.all_albums:
            print(album.title)
        
    @classmethod
    def get_album_from_name(cls,name):
        for album in cls.all_albums:
            if album.name==name:
                return album
        print("get_album_from_name: Object not found")
        return None
    
    @classmethod    
    def display_album_from_artist_name(cls,artist_name):
        for album in cls.all_albums:
            if album.artist.name==artist_name:
                print(album.title)
        
        
####################################################################            
class Track:
    all_tracks=[]
    
    def __init__(self,track_id,name,album):
        self.id=track_id
        self.name=name
        self.album=album
        
    @classmethod
    def loads_all_tracks(cls,cursor):
        cursor.execute('''SELECT tracks.TrackId, tracks.Name,tracks.AlbumId FROM tracks''')
        result=cursor.fetchall()

        for record in result:
            album=Album.get_album_object_from_id(record[2])
            track=cls(record[0],record[1],album)
            cls.all_tracks.append(track)
            
    @classmethod
    def display_tracks_from_album_name(cls,name):
        for track in cls.all_tracks:
            if track.album!=None and track.album.title==name:
                print(track.name) 
                
    @classmethod
    def display_tracks_from_artist_name(cls,artist_name):
        for track in cls.all_tracks:
            if track.album!=None and track.album.artist.name==artist_name:
                print(track.name) 
####################################################################  
# MAIN
db = sqlite3.connect('chinook.db')
cursor = db.cursor()
Artist.loads_all_artists(cursor)
Album.loads_all_albums(cursor)
Track.loads_all_tracks(cursor)

#print(Artist.all_artists)
#print(Album.all_albums)
#print(Track.all_tracks)

#Album.display_all_albums()

#Track.display_tracks_from_album_name("Live After Death")

#Album.display_album_from_artist_name("Jamiroquai")

Track.display_tracks_from_artist_name("Jamiroquai")


```
