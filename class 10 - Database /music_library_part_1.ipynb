```python
import sqlite3
db = sqlite3.connect('chinook.db')
cursor = db.cursor()
cursor.execute('''SELECT artists.Name,albums.Title 
                FROM artists,albums
                WHERE artists.ArtistId=albums.ArtistId''')
result=cursor.fetchall()

for record in result:
    print("{} {}".format(record[0],record[1]))
```
