```python
import sqlite3
####################################################################  
class Customer:

    def __init__(self,customer_id,first_name,last_name):
        self.customer_id=customer_id
        self.first_name=first_name
        self.last_name=last_name
    
    @classmethod
    def load_customer(cls,first_name,last_name,cursor):
        cursor.execute('''SELECT CustomerId
                          FROM customers 
                          WHERE FirstName=?
                          AND LastName=?''',(first_name,last_name))
        
        result=cursor.fetchone()
        if result!=None:           
            return Customer(result[0],first_name,last_name)
        else:
            return None
####################################################################  
class Order:

    def __init__(self,order_id, order_date, billing_address, billing_city,list_items):
        self.order_id=order_id
        self.order_date=order_date
        self.billing_address=billing_address
        self.billing_city=billing_city
        self.list_items=list_items
    
    @classmethod
    def load_orders(cls,customer_id,cursor):
            list_order=[]
            # Get all invoices for a specific customer
            cursor.execute('''SELECT InvoiceId, InvoiceDate, BillingAddress, BillingCity
                              FROM invoices
                              WHERE CustomerId=?
                            ''',(customer_id,))
            orders=cursor.fetchall()
            
            # For each invoice (for this specific customer) get all related items
            for order in orders:
                cursor.execute('''SELECT 
                                  inv.UnitPrice, inv.Quantity,
                                  tra.Name,alb.Title,art.Name
                                  FROM invoice_items inv, 
                                  (tracks tra LEFT JOIN albums alb ON tra.AlbumId=alb.AlbumId), 
                                  artists art
                                  WHERE inv.TrackId=tra.TrackId
                                  AND alb.ArtistId=art.ArtistId
                                  AND inv.InvoiceId=?''',(order[0],))
                
                
                order_items=cursor.fetchall()
                items_list=[]
                for item in order_items:
                    item={'unit_price':item[0],
                    'quantity':item[1],
                    'song':item[2],
                    'album':item[3],
                    'artist':item[4]}
                    items_list.append(item)
                    
               #Build a list of objects orders
                #InvoiceId, InvoiceDate, BillingAddress, BillingCity,list of items
                order=Order(order[0],order[1],order[2],order[3],items_list)
                list_order.append(order)
            return list_order
                
            
            

        
####################################################################  
# MAIN
db = sqlite3.connect('chinook.db')
cursor = db.cursor()

login=(input("Insert Login:")).capitalize()
password=(input("Insert Password:")).capitalize()
customer=Customer.load_customer(login,password,cursor)
number_try=2

while(customer==None and number_try>0):
    number_try-=1
    print("This login and password is not correct.")
    login=(input("Insert Login:")).capitalize()
    password=(input("Insert Password:")).capitalize()
    customer=Customer.load_customer(login,password,cursor)
    

if number_try>0:    
    print("You are connected")
    orders=Order.load_orders(customer.customer_id,cursor)
    for order in orders:
        print("{} {}".format(order.order_id,order.order_date))
        for item in order.list_items:
            print("\t| {} \t\t| {} \t| {} \t| {} \t|".format(item["song"],item["artist"],item["quantity"],item["unit_price"]))
    
else:
    print("You are not connected")
    
    
    




```
